```python
## Generate Traits

TOTAL_IMAGES = 100 # Number of random unique images we want to generate

all_images = [] 

# A recursive function to generate unique image combinations
def create_new_image():
    
    new_image = {} #

    # For each trait category, select a random trait based on the weightings 
    new_image["Background"] = random.choices(backgrounds, backgroundweights)[0]
    new_image["Martian Type"] = random.choices(martianbodys, martianbodyweights)[0]
    new_image["Head"] = random.choices(head, headweights)[0]
    new_image["Eyes"] = random.choices(eyes, eyesweights)[0]
    new_image["Mouth"] = random.choices(mouth, mouthweights)[0]
    new_image["Costume"] = random.choices(costume, costumeweights)[0]
    new_image["Accessories"] =random.choices(accessories, accessoriesweights)[0]
    new_image["Shine"] = random.choices(shine, shineweights)[0]

    if new_image in all_images:
        return create_new_image()
    else:
        return new_image
    
    
# Generate the unique combinations based on trait weightings
for i in range(TOTAL_IMAGES): 
    
    new_trait_image = create_new_image()
    
    all_images.append(new_trait_image)
    
```

```python
# Returns true if all images are unique
def all_images_unique(all_images):
    seen = list()
    return not any(i in seen or seen.append(i) for i in all_images)

print("Are all images unique?", all_images_unique(all_images))
```

```python
  
    
#### Generate Images    
for item in all_images:

    im1 = Image.open(f'./Backgrounds/{backgroundfiles[item["Background"]]}.png').convert('RGBA')
    im2 = Image.open(f'./Martians/{martianbodyfiles[item["Martian Type"]]}.png').convert('RGBA')
    im3 = Image.open(f'./Heads/{headfiles[item["Head"]]}.png').convert('RGBA')
    im4 = Image.open(f'./Eyes/{eyefiles[item["Eyes"]]}.png').convert('RGBA')
    im5 = Image.open(f'./Mouths/{mouthfiles[item["Mouth"]]}.png').convert('RGBA')
    im6 = Image.open(f'./Costumes/{costumefiles[item["Costume"]]}.png').convert('RGBA')
    im7 = Image.open(f'./Accessories/{accessoriesfiles[item["Accessories"]]}.png').convert('RGBA')
    im8 = Image.open(f'./Shine/{shinefiles[item["Shine"]]}.png').convert('RGBA')


    #Create each composite
    com1 = Image.alpha_composite(im1,  im2)
    com2 = Image.alpha_composite(com1, im3)
    com3 = Image.alpha_composite(com2, im4)
    com4 = Image.alpha_composite(com3, im5)
    com5 = Image.alpha_composite(com4, im6)
    com6 = Image.alpha_composite(com5, im7)
    com7 = Image.alpha_composite(com6, im8)

   
    
    #Convert to RGB
    rgb_im = com7.convert('RGBA')
    display(rgb_im.resize((128,128), Image.NEAREST))

    file_name = str(item["tokenId"]) + ".png"
    rgb_im.save("./output/" + file_name)
    print(f'{str(item["tokenId"])} done')
    
```

```python
#### Generate Metadata for each Image    

f = open('./metadata/all-traits.json',) 
data = json.load(f)


IMAGES_BASE_URI = ""
PROJECT_NAME = "UrbanMartians"

def getAttribute(key, value):
    return {
        "trait_type": key,
        "value": value
    }
for i in data:
    token_id = i['tokenId']
    token = {
        "image": IMAGES_BASE_URI + str(token_id) + '.png',
        "tokenId": token_id,
        "name": PROJECT_NAME + '#' + str(token_id),
        "external_url": "https://urbanmartians.net",
        "description": "UrbanMartians are 3000 uniquely generated NFT's on the Ethereum Blockchain",
        "attributes": []
    }
    token["attributes"].append(getAttribute("Background", i["Background"]))
    token["attributes"].append(getAttribute("Martian Type", i["Martian Type"]))
    token["attributes"].append(getAttribute("Head", i["Head"]))
    token["attributes"].append(getAttribute("Eyes", i["Eyes"]))
    token["attributes"].append(getAttribute("Mouth", i["Mouth"]))
    token["attributes"].append(getAttribute("Costume", i["Costume"]))
    token["attributes"].append(getAttribute("Accessories", i["Accessories"]))
    token["attributes"].append(getAttribute("Shine", i["Shine"]))


    with open('./metadata/' + str(token_id), 'w') as outfile:
        json.dump(token, outfile, indent=4)
f.close()
```
